```python
def pearson_corr(x, y):
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    numerator = sum((x - mean_x) * (y - mean_y))
    denominator = np.sqrt(sum((x - mean_x)**2) * sum((y - mean_y)**2))
    if denominator == 0:
        return 0  # Return 0 to avoid division by zero
    else:
        return numerator / denominator
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity


# Function to get recommendations
def get_recommendations(user_id, df):
    # Compute similarity matrix using cosine similarity
    similarity_matrix = cosine_similarity(df.T)

    # Get similarity scores for the given user
    user_similarity = similarity_matrix[user_id]

    # Get items rated by the user
    rated_items = df.iloc[:, user_id]

    # Initialize recommendation list
    recommendations = {}

    # Iterate over each user
    for user_index, user_score in enumerate(user_similarity):
        # Skip the current user
        if user_index == user_id:
            continue

        # Get items rated by the current user
        user_items = df.iloc[:, user_index]

        # Iterate over each item
        for item_index, item_score in enumerate(user_items):
            # Check if the item is not rated by the target user and the user has some similarity
            if rated_items[item_index] == 0 and item_score > 0:
                # Add the item to recommendations with its similarity score
                if item_index not in recommendations:
                    recommendations[item_index] = 0
                recommendations[item_index] += item_score * user_score

    # Sort recommendations by score
    recommendations = sorted(recommendations.items(), key=lambda x: x[1], reverse=True)

    return recommendations

```

```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
```

```python
movieRating_feature = newdf2.pivot_table(index='name_y',columns='user_id',values='rating').fillna(0)
```

```python
def recommendedMovie(movieRating_feature):
    mat_movie_features_array = csr_matrix(movieRating_feature.values)
    model_knn = NearestNeighbors(metric='cosine', algorithm='brute',n_jobs=-1)
    model_knn.fit(mat_movie_features_array)
    query_index =np.random.choice(movieRating_feature.shape[0])
    print(query_index)
    distances,indices =model_knn.kneighbors(movieRating_feature.iloc[query_index,:].values.reshape(1,-1), n_neighbors=6)
    print("Distances -->",distances," Indices -->",indices)
 
    print(distances.flatten())
    print(len(distances.flatten()))
 
    for i in range(0, len(distances.flatten())):
        if i == 0:
            print('Recommendations for {0}:\n'.format(movieRating_feature.index[query_index]))
        else:
            print('{0}: {1}, with distance of {2}:'.format(i, movieRating_feature.index[indices.flatten()[i]], distances.flatten()[i]))
```
